```python
# Initialize new columns for Program and Starting_Semester
df = pd.read_csv("..\inputs_new\Graduate Study Plans -revised(Sheet1).csv",header=1)
df = df[['Unnamed: 0', 'Fall ', 'Spring*', 'Summer ', 'Fall', 'Spring*.1']]
df['Program'] = None
df['Starting_Semester'] = None

current_program = None
current_semester = None
section = None
# Iterate through the rows of the dataframe to populate the new columns
drop_indexes = []
for index, row in df.iterrows():
    # is section is none or changed then drop that index

    section_now = row['Unnamed: 0']
    section = section_now
    if pd.notna(section):
        if '-' in section:
            current_program = section.strip()
            drop_indexes.append(index)
        elif 'Start' in section:
            current_semester = section.split()[0]  # Extract 'Fall' or 'Spring'
    df.at[index, 'Program'] = current_program
    df.at[index, 'Starting_Semester'] = current_semester

    df.at[index, 'drop_index'] = True if index in drop_indexes else False
    


df = df.drop(drop_indexes)
print(df.columns)
# select following columsn and rename properly
# 'Fall ', 'Spring*', 'Summer ', 'Fall', 'Spring*.1','Program', 'Starting_Semester'],
df = df[['Fall ', 'Spring*', 'Summer ', 'Fall', 'Spring*.1','Program', 'Starting_Semester']]
df.columns = ['Fall1', 'Spring1', 'Summer1', 'Fall2', 'Spring2', 'Program', 'Starting_Semester']
# Display the updated dataframe
display(df)
```
